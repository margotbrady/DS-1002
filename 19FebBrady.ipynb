```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
import requests

# As an example, I have been using 'https://www.peruforless.com/blog/wp-content/uploads/2017/06/rainbow-mountain-peru.jpg' or
# 'https://upload.wikimedia.org/wikipedia/commons/thumb/5/5c/Double-alaskan-rainbow.jpg/550px-Double-alaskan-rainbow.jpg' because they are particularly colorful.


def showimage():
  url = input("This is a color filter creator. Please insert an image url you would like to use: " +"\n")
  img=np.asarray(Image.open(requests.get(url,stream=True).raw))
  print("This is your original image:")
  plt.imshow(img)
  plt.show()

  print("This is your red-filtered image:")
  red_version = np.copy(img)
  red_version[:,:,1:]=0
  plt.imshow(red_version)
  plt.show()

  print("This is your green-filtered image:")
  green_version=np.copy(img)
  green_version[:,:,::2]=0
  plt.imshow(green_version)
  plt.show()

  print("This is your blue-filtered image:")
  blue_version=np.copy(img)
  blue_version[:,:,:2]=0
  plt.imshow(blue_version)
  plt.show()

  print("This is your image with both R and G values but no blue.")
  rgnoblue=np.copy(img)
  rgnoblue[:,:,2:]=0
  plt.imshow(rgnoblue)
  plt.show()

  print("This is your image with both R and B values but no green.")
  rbnogreen=np.copy(img)
  rbnogreen[:,:,1:2]=0
  plt.imshow(rbnogreen)
  plt.show()

  print("This is your image with both G and B values but no red.")
  gbnored=np.copy(img)
  gbnored[:,:,:1]=0
  plt.imshow(gbnored)
  plt.show()

  # Here we go back and recombine all of the versions together. I use np.stack to combine the versions along the same axis.
  print("If we combine all of the versions back togethere again, we get the original version:")
  imgcombined=np.stack((red_version[:,:,0],green_version[:,:,1],blue_version[:,:,2]),axis=2)
  plt.imshow(imgcombined)
  plt.show()
  print("We can test if these images are the same by comparing them to each other. False means they are not the same array, while True means they are.")
  print(imgcombined==img)

showimage()






```
